```python
import matplotlib.pyplot as plt
import pickle

# show gpu info colab
# gpu_info = !nvidia-smi
# gpu_info = '\n'.join(gpu_info)
# if gpu_info.find('failed') >= 0:
#  print('Not connected to a GPU')
# else:
#   print(gpu_info)

# load history
history_custom_cnn = pickle.load(open(f"custom_cnn_300_history.sav", "rb"))
history_vgg16 = pickle.load(open(f"pretrained_vgg16_500_history.sav", "rb"))
history_vgg16_valtest = pickle.load(open(f"pretrained_vgg16_valtest_200_history.sav", "rb"))
history_resnet50 = pickle.load(open(f"pretrained_restnet50_20_history.sav", "rb"))

plt.figure(figsize=(12.0, 7.0))
plt.xlabel('Epoch')

plt.ylabel('Loss')

plt.plot(history_custom_cnn['loss'], color='blue', label='Custom CNN Train')
plt.plot(history_custom_cnn['val_loss'], color='lightsteelblue', label='Custom CNN Validation')
plt.plot(history_vgg16['loss'], color='red', label='VGGNet 16 Train')
plt.plot(history_vgg16['val_loss'], color='indianred', label='VGGNet 16 Validation')
plt.plot(history_vgg16_valtest['loss'], color='black', label='VGGNet 16 val=test Train')
#plt.plot(history_vgg16_valtest['val_loss'], color='darkgray', label='VGGNet 16 val=test Validation')
plt.plot(history_resnet50['loss'], color='darkgreen', label='ResNet 50 Train')
plt.plot(history_resnet50['val_loss'], color='greenyellow', label='ResNet 50 Validation')
#plt.xlim([0, 200])
#plt.xlim([0, 20])


plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
